```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display


df = pd.read_excel('power_data.xlsx')

#  Calculate Power and Energy
df['Power (W)'] = df['Voltage (V)'] * df['Current (A)']
df['Energy (Wh)'] = df['Power (W)'] * df['Time Used (hr)']


display(df)

#  Plot 1: Power Consumption by Device
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Device', y='Power (W)', palette='viridis')
plt.title('Power Consumption by Device')
plt.ylabel('Power (Watts)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#  Plot 2: Total Energy Consumption
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Device', y='Energy (Wh)', palette='magma')
plt.title('Total Energy Consumption')
plt.ylabel('Energy (Wh)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#  Plot 3: Pie Chart - Energy Share by Device
plt.figure(figsize=(6,6))
plt.pie(df['Energy (Wh)'], labels=df['Device'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Energy Consumption Share')
plt.show()

```
